### Folium을 활용한 지도 그리기

```python
import folium
```

```python
m = folium.Map(location = [37.5203885,127.0111537], zoom_start = 20) # 경도위도, 줌
folium.Marker(location = [37.5203885,127.0111537], popup="위치1").add_to(m)
m
```
